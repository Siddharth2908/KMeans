# 7. Provide an analysis of trends in gender ratio against region population:
    7.1 Plot the gender ratio as a function of the population of the region.
    7.2 Comment on any trends you see in the data. What could cause such trends?

```r
Ratio <- table(Source$gender)
Ratio
```

```r
Source_Male <- filter(Source, gender == "M")
Source_Female <- filter(Source, gender == "F")

Source_Male_Summary <- Source_Male %>%group_by(region)%>% summarise(pop=sum(population))

Source_Female_Summary <- Source_Female %>%group_by(region)%>% summarise(pop=sum(population))

Male_Female_Ratio <- inner_join(Source_Male_Summary,Source_Female_Summary,by="region")

Male_Female_Ratio$Male_Female_ratio <- (Male_Female_Ratio$pop.x/Male_Female_Ratio$pop.y)
Male_Female_Ratio$total_pop <- (Male_Female_Ratio$pop.x+Male_Female_Ratio$pop.y)

Male_Female_Ratio_Region <- subset(Male_Female_Ratio , select = -c(pop.x,pop.y))

Male_new_ratio_plot <- ggplot(Male_Female_Ratio_Region,aes(x=total_pop)) + 
  geom_point(aes(y = Male_Female_ratio), color = "red") +
  xlab("Total Population") +
  ylab("Male:Female ratio")+
  ggtitle("Comparing Gender to Population")+
  coord_cartesian(xlim = c(0, 1000), ylim = c(0, 6))

Male_new_ratio_plot
```

Gender ratio is equal i.e 50% Male, 50% Female
